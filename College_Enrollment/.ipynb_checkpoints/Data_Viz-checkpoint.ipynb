### `Class: Data`
### Functions: `__init__()`, `variables()`, `line_plot()`, `scatter_plot()`
---
>#### First, we clean the data by removing the columns that don't contain any data. In this dataset, the original authors used empty columns as vertical dividers to better visualize their data. For our purposes, we must delete them.
>#### Second, we must note that some indexes are used as column headers. Creating a new dataframe, or a series object is helpful.
>#### Third, functions are created to deal with creating figures.

```python
import pandas as pd
import matplotlib.pyplot as plt
class Data:
    def __init__(self):
        # performs the necessary clean-ups.
        self.data = pd.read_excel('College-enrollment-rates-trans_verified.xlsx')
        self.data.dropna()
        self.data.drop(columns=['Unnamed: 2', 'Unnamed: 4', 'Unnamed: 6', 'Unnamed: 8',
                           'Unnamed: 10', 'Unnamed: 12', 'Unnamed: 14', 'Unnamed: 16' ,
                           'Unnamed: 18', 'Unnamed: 20', 'Unnamed: 22', 'Unnamed: 24',
                          'Unnamed: 26'], index=[29,30,31,32,33,34], inplace=True)
        self.data.rename(columns={'Unnamed: 1': '2000', 'Unnamed: 3': '2001', 'Unnamed: 5': '2002', 'Unnamed: 7': '2003',
                           'Unnamed: 9': '2004', 'Unnamed: 11': '2005', 'Unnamed: 13': '2006', 'Unnamed: 15': '2007' ,
                           'Unnamed: 17': '2008', 'Unnamed: 19': '2009', 'Unnamed: 21': '2010', 'Unnamed: 23': '2011',
                          'Unnamed: 25': '2012'}, inplace=True)
        # attempts to convert data types to floating points.
        self.data.convert_dtypes()
        self.data['2000'] = pd.to_numeric(self.data['2000'], errors="coerce")
        self.data['2001'] = pd.to_numeric(self.data['2001'], errors="coerce")
        self.data['2002'] = pd.to_numeric(self.data['2002'], errors="coerce")
        self.data['2008'] = pd.to_numeric(self.data['2008'], errors="coerce")

    def variables(self):
        # Gathers the sprecific data we need using pandas
        # We create a new DataFrame from the index we choose.
        self.Male_White = pd.DataFrame(self.data.loc[14])
        self.Female_White = pd.DataFrame(self.data.loc[22])
        self.Male_Black = pd.DataFrame(self.data.loc[15])
        self.Female_Black = pd.DataFrame(self.data.loc[23])
        self.Male_Hispanic = pd.DataFrame(self.data.loc[16])
        self.Female_Hispanic = pd.DataFrame(self.data.loc[24])
        self.Male_Asian = pd.DataFrame(self.data.loc[17])
        self.Female_Asian = pd.DataFrame(self.data.loc[25])

        # We must drop the sub-heading used in the original dataframe. Remember that this subheading is an index.
        self.Male_White.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)
        self.Female_White.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)
        self.Male_Black.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)
        self.Female_Black.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)
        self.Male_Hispanic.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)
        self.Female_Hispanic.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)
        self.Male_Asian.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)
        self.Female_Asian.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)

        self.data_list = [self.Male_White, self.Female_White,
                         self.Male_Black, self.Female_Black,
                         self.Male_Asian, self.Female_Asian]
        self.titles = ['White Males', 'White Females',
                       'Black Males', 'Black Females',
                       'Asian Males', 'Asian Females']
        
    def line_plot(self):
        # Creates the plots we need for Data Visualization
        fig, axes = plt.subplots(1,3, figsize=(12,4))

        with pd.plotting.plot_params.use("x_compat", True):
            self.Male_White.plot(ax=axes[0], title="White", color='blue', marker="+", ylim=(0,100))
            self.Female_White.plot(ax=axes[0], color='red', marker="+", ylim=(0,100))
            axes[0].legend(['Male', 'Female'])
        with pd.plotting.plot_params.use("x_compat", True):
            self.Male_Black.plot(ax=axes[1], title="Black", color='blue', marker="+", ylim=(0,100))
            self.Female_Black.plot(ax=axes[1], color='red', marker="+", ylim=(0,100))
            axes[1].legend(['Male', 'Female'])
        with pd.plotting.plot_params.use("x_compat", True):
            self.Male_Asian.plot(ax=axes[2], title="Asian", color='blue', marker="+", ylim=(0,100))
            self.Female_Asian.plot(ax=axes[2], color='red', marker="+", ylim=(0,100))
            axes[2].legend(['Male', 'Female'])

        plt.suptitle('Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by gender & ethnicity, 2000–2012')
        plt.show()

    def scatter_plot(self):
        # Creates the plots we need for Data Visualization
        plt.figure()
        with pd.plotting.plot_params.use("x_compat", True):
            plt.scatter(x=self.Male_White[14].index, y=self.Male_White[14].values, label="White Males", color='blue')
            plt.scatter(x=self.Female_White[22].index, y=self.Female_White[22].values, label="White Females", color='red')
            
        plt.legend()
        plt.suptitle('Percentage of White 18- to 24-year-olds enrolled in 2- and 4-year colleges by gender, 2000–2012')
        plt.show()
```

```python
Data = Data()
Data.variables()
Data.line_plot()
```

> ###Insights for White, Black, and Asian groups
> ####  Enrollment is overall going in a downward trend. Asian males lead in enrollment, White males are the median enrolled group, while Black males are in the lowest rate of enrollment; however they are the only group that has a positive trend-albiet slow.
> #### For females, enrollment is also slow. Asian females lead in enrollment, White females are in the median, and Black females are in the lowest group. Both White and Black groups are seeing a downward trend.
