### `Class: Data`
### Functions: `__init__()`, `variables()`, `line_plot()`, `scatter_plot()`
---
>#### First, we clean the data by removing the columns that don't contain any data. In this dataset, the original authors used empty columns as vertical dividers to better visualize their data. For our purposes, we must delete them.
>#### Second, we must note that some indexes are used as column headers. Creating a new dataframe, or a series object is helpful.
>#### Third, functions are created to deal with creating figures.

```python
import pandas as pd
import matplotlib.pyplot as plt
class Data:
    def __init__(self):
        # performs the necessary clean-ups.
        self.data = pd.read_excel('College-enrollment-rates-trans_verified.xlsx')
        self.data.dropna()
        self.data.drop(columns=['Unnamed: 2', 'Unnamed: 4', 'Unnamed: 6', 'Unnamed: 8',
                           'Unnamed: 10', 'Unnamed: 12', 'Unnamed: 14', 'Unnamed: 16' ,
                           'Unnamed: 18', 'Unnamed: 20', 'Unnamed: 22', 'Unnamed: 24',
                          'Unnamed: 26'], index=[29,30,31,32,33,34], inplace=True)
        self.data.rename(columns={'Unnamed: 1': '2000', 'Unnamed: 3': '2001', 'Unnamed: 5': '2002', 'Unnamed: 7': '2003',
                           'Unnamed: 9': '2004', 'Unnamed: 11': '2005', 'Unnamed: 13': '2006', 'Unnamed: 15': '2007' ,
                           'Unnamed: 17': '2008', 'Unnamed: 19': '2009', 'Unnamed: 21': '2010', 'Unnamed: 23': '2011',
                          'Unnamed: 25': '2012'}, inplace=True)
        # attempts to convert data types to floating points.
        self.data.convert_dtypes()
        self.data['2000'] = pd.to_numeric(self.data['2000'], errors="coerce")
        self.data['2001'] = pd.to_numeric(self.data['2001'], errors="coerce")
        self.data['2002'] = pd.to_numeric(self.data['2002'], errors="coerce")
        self.data['2008'] = pd.to_numeric(self.data['2008'], errors="coerce")

    def variables(self):
        # Gathers the sprecific data we need using pandas
        # We create a new DataFrame from the index we choose.
        self.Male_White = pd.DataFrame(self.data.loc[14])
        self.Female_White = pd.DataFrame(self.data.loc[22])

        # We must drop the sub-heading used in the original dataframe. Remember that this subheading is an index.
        self.Male_White.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)
        self.Female_White.drop(index='Percentage of 18- to 24-year-olds enrolled in 2- and 4-year colleges by sex and race/ethnicity, 2000–2012', inplace=True)

        
    def line_plot(self):
        # Creates the plots we need for Data Visualization
        figure_1 = plt.figure()
        plt.plot(self.Male_White, label="White Males", color='blue')
        plt.title('White Males')
        plt.legend()
        
        figure_2 = plt.figure()
        plt.plot(self.Female_White, label="White Females", color='red')
        plt.title('White Females')
        plt.legend()
        
        plt.show()

    def scatter_plot(self):
        figure_1 = plt.figure()
        plt.scatter(x=self.Male_White[14].index, y=self.Male_White[14].values, label="White Males", color='blue')
        plt.title('White Males')
        plt.legend()

        figure_2 = plt.figure()
        plt.scatter(x=self.Female_White[22].index, y=self.Female_White[22].values, label="White Females", color='red')
        plt.title('White Males')
        plt.legend()

        plt.show()
```

```python
Data = Data()
Data.variables()
Data.line_plot()
```
